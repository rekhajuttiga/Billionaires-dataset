```python
# Remove these unnecessary columns: Rank, lastname, firstname(not usefull for analysis), birthDate (birth date info is also included in separate columns), date (not useful for analysis), source (not necessary for analysis because there are too many unique values and what is needed is the industry, not the specific business name or product), and category (contains identical information as 'industry' field).\n",
```

```python
# dropping unneccessery columns 

df = df.drop(columns = ['birthDate', 'date', 'rank', 'lastName', 'firstName', 'category'])
```

```python
# Investigating Missing Values
    

df.isnull().sum()
```

#There are too many missing values in the 'organization' and 'title' columns so these will not helps for analysis. These columns will be removed. The other missing values in the dataset are cant be imputed and removed as the data contains information of different countries and if remove them it will be inaccurate.

```python
df = df.drop(columns = (['organization', 'title']))
```

```python
df.shape
```

```python
# Looking for mixed data types
    
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
     print (col)
   
```

```python
df.dtypes
```

```python
# These columns with mixed data types are all strings with some data missing. I will not change the data types on these columns to make them completely strings because when doing that, the missing values no longer appear as missing when running this script.  
## The data that is there is the correct data type except 'gdp_country'. That field needs to be converted to int64 or float64.
```

```python
# Removing dollar signs from gdp_country column and changing data
    
df['gdp_country'] = df['gdp_country'].replace('[\\$,]', '', regex=True).astype(float)
```
